### **Observation Set : 5**
> There are `8829` values in `M` unit

> There are `316` values in `k` unit

> There are `1695` values in `Varies with device`.

> Now its time to convert the above units into bytes and we will remove the `M` and `k` from the values. Also, convert to them into numerical datatype for further processing.

```python
# Let's convert the size column into numeric
#  This function will convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it

def convert_size(Size):
    # add function details here
 
    if isinstance(Size, str):
        if 'k' in Size:
            return float(Size.replace('k', '')) * 1024
        elif 'M' in Size:
            return float(Size.replace('M', '')) * 1024 * 1024
        elif 'Varies with device' in Size:
            return np.nan
    return Size

df['Size'] = df['Size'].apply(convert_size)
```

```python
# rename the column name 'Size' to 'Size_in_bytes'
df.rename(columns={'Size': 'Size_in_bytes'}, inplace=True)
```

```python
# Let's remove the rows which have missing values
df.dropna(subset=['Current Ver', 'Android Ver', 'Category', 'Type', 'Genres'], inplace=True)
```

```python
# Group by 'Installs_category' and calculate the mean rating for each group
mean_ratings = df.groupby('Installs_category')['Rating'].transform('mean')
# Fill missing values in 'Rating' with the respective group mean
df['Rating'] = df['Rating'].fillna(mean_ratings)
```

```python
# Calculate median sizes by app category
median_sizes = df.groupby('Category')['Size_in_bytes'].transform('median')
# Impute missing values with median sizes
df['Size_in_bytes'].fillna(median_sizes, inplace=True)
# Impute missing values with median sizes
df['Size_in_Mb'].fillna(median_sizes, inplace=True)
```

### **Observations set: 11**

>The rows with missing values in the `Current Ver`, `Android Ver`, `Category`, `Type`, and `Genres` columns were removed, ensuring no null values in these critical categorical columns.
  
>The missing values in the `Rating` column were imputed with the mean rating of their respective `Installs_category` groups, contextualizing the imputation to maintain meaningful data relationships.

>Missing values in the `Size_in_bytes` and `Size_in_Mb` columns were imputed with category-specific median sizes, ensuring data completeness and preserving contextual relevance for app size analysis
